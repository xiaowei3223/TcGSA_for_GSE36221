# TcGSA

## gmt object

```r
kegg_gmt = GSA::GSA.read.gmt("c2.cp.kegg.v7.0.symbols.gmt")
```

## Prepare Design

```r
Design_TcGSA_tre012 = pdata_de[c(2,38,39,40)]
names(Design_TcGSA_tre012) = c("Sample_ID", "PatientID", "TimePoint", "Treatment")
Design_TcGSA_tre012$TimePoint = as.numeric(Design_TcGSA_tre012$TimePoint)
Design_TcGSA_tre012$PatientID = as.factor(Design_TcGSA_tre012$PatientID)
Design_TcGSA_tre012$Treatment = as.factor(Design_TcGSA_tre012$Treatment)
```

```r
head(Design_TcGSA_tre012)
dim(Design_TcGSA_tre012)
```

## expression data and related design and TcGSA test

In this part codes, we use loop circle to perform multiple comparing. At the first, one group run  `TcGSA::TcGSA.LR`,then,runs by comaparing treatment:

Varible | comparing
--------:|:---------------
tre01| treatment 0 VS treatment 1
tre02|treatment 0 VS treatment 2
tre12|treatment 1 VS treatment 2
tre012|treatment 0 VS treatment 1 VS treatment 2

what it means for number 0, 1 and 2:

number | treatment
---------:|:------------
0 | placebo 30 months
1 | fluticasone 30 months
2 | fluticasone/ salmeterol 30 months

```r
head(GSE36221)
dim(GSE36221)
```

```r
TcGSA_result = list()
TcGSA_result_pvalues = list()
TcGSA_expr = list()
TcGSA_design = list()
for (i in 1:7){
  if (i >3 ){
      #for comparing three condition about time course data
    if (i == 4){Design_TcGSA = Design_TcGSA_tre012}
      #for comparing two condition about time course data
    if (i >4){Design_TcGSA = Design_TcGSA_tre012[which(Design_TcGSA_tre012$Treatment != (i-5)),]}
    expr = GSE36221[,Design_TcGSA$Sample_ID]
    tcgsa_result0 <- TcGSA::TcGSA.LR(expr =expr, 
                                     gmt = kegg_gmt,
                                     design = Design_TcGSA, 
                                     subject_name = "PatientID", 
                                     time_name = "TimePoint", 
                                     group_name = "Treatment")}
  else{  
      #for one condition time course data
    Design_TcGSA = Design_TcGSA_tre012[which(Design_TcGSA_tre012$Treatment ==(i-1)),] 
    expr = GSE36221[,Design_TcGSA$Sample_ID]
    tcgsa_result0 <- TcGSA::TcGSA.LR(expr =expr, 
                                       gmt = kegg_gmt,
                                       design = Design_TcGSA, 
                                       subject_name = "PatientID", 
                                       time_name = "TimePoint")}
  
  TcGSA_expr[[i]] <- expr
  TcGSA_design[[i]] <- Design_TcGSA
  TcGSA_result[[i]] <- tcgsa_result0
  TcGSA_result_pvalues[[i]] <- TcGSA::signifLRT.TcGSA(tcgsa_result0)$mixedLRTadjRes
}

names(TcGSA_expr) = c('tre0','tre1','tre2', 'tre012', 'tre12','tre02', 'tre01')
names(TcGSA_design) = c('tre0','tre1','tre2', 'tre012', 'tre12','tre02', 'tre01')
names(TcGSA_result) = c('tre0','tre1','tre2', 'tre012', 'tre12','tre02', 'tre01')
names(TcGSA_result_pvalues) = c('tre0','tre1','tre2', 'tre012', 'tre12','tre02', 'tre01')
```

# heatmap

make a matrix for summary all p values in each comparing varibles.

```r
TcGSA_result_pvalues0 <- TcGSA_result_pvalues
#change the Adj.pvalue column names
for (i in 1:length(TcGSA_result_pvalues)){
  names(TcGSA_result_pvalues[[i]])[2] = names(TcGSA_result_pvalues)[i]
}
#make all p values in a dataframe
allPvalue = TcGSA_result_pvalues$tre012[,1:2]
for (i in c("tre12" , "tre02" , "tre01")){
  allPvalue = merge(allPvalue,TcGSA_result_pvalues[[i]][,1:2], by = "GeneSet", all = TRUE)
}
#row names are geneset and remove GeneSet columns in allPvalue
RowNames = allPvalue$GeneSet
allPvalue = allPvalue[-1]
allPvalue = apply(allPvalue , 2, as.numeric)
rownames(allPvalue) = RowNames
```
